### Cleaning Data

```python
kolom_outliers = ['humidity', 'windspeed', 'casual']

for kolom in kolom_outliers:
    Q1 = df[kolom].quantile(0.25)
    Q3 = df[kolom].quantile(0.75)

    # Menghitung IQR
    IQR = Q3 - Q1


    maximum = Q3 + (1.5 * IQR)
    minimum = Q1 - (1.5 * IQR)

    kondisi_lower_than = df[kolom] < minimum
    kondisi_more_than = df[kolom] > maximum


    df[kolom].mask(cond=kondisi_more_than, other=maximum, inplace=True)
    df[kolom].mask(cond=kondisi_lower_than, other=minimum, inplace=True)

# Menampilkan dataframe setelah outlier ditangani
print(df[kolom_outliers].head())
```

**Insight:**
- Data menunjukkan nilai `humidity` yang berkisar antara 0.437273 hingga 0.805833, sementara `windspeed` berada pada rentang 0.160296 hingga 0.248539. Nilai-nilai ini mencerminkan stabilitas dalam kondisi kelembaban dan kecepatan angin di area yang diteliti. Dengan menangani outlier, nilai-nilai ini menjadi lebih representatif, yang memungkinkan analisis yang lebih akurat mengenai hubungan antara kelembaban dan kecepatan angin terhadap faktor-faktor lain seperti jumlah pengguna atau total penggunaan. Data yang bersih dan tidak terpengaruh oleh outlier memberikan gambaran yang lebih jelas tentang kondisi lingkungan yang mungkin mempengaruhi perilaku pengguna.

- Kolom `casual`, yang menunjukkan jumlah pengguna yang terdaftar, menunjukkan nilai antara 82.0 hingga 331.0 setelah penanganan outlier. Rentang ini memberikan wawasan mengenai variasi penggunaan yang terjadi di luar faktor-faktor eksternal yang mungkin mendistorsi data. Dengan mengganti outlier, analisis lebih lanjut dapat dilakukan untuk memahami pola penggunaan, seperti waktu tertentu dalam minggu atau bulan ketika jumlah pengguna cenderung meningkat. ### Pertanyaan 1
Bagaimana pengaruh musim (season) terhadap jumlah total pengguna (total_user) pada setiap bulan (month) selama periode analisis?

```python
season_month = df.groupby(['season', 'month'])['total_users'].sum().reset_index()
```

```python
monthly_users = season_month.groupby(['month', 'season']).agg({'total_users': 'sum'}).reset_index()

# Menghitung koefisien korelasi
correlation = monthly_users['total_users'].corr(monthly_users['season'].astype('category').cat.codes)

# Visualisasi dengan scatter plot (bubble chart)
plt.figure(figsize=(10, 6))

# Menyimpan ukuran untuk legend
legend_handles = []

for season in monthly_users['season'].unique():
    subset = monthly_users[monthly_users['season'] == season]
    # Membuat bubble chart
    scatter = plt.scatter(subset['month'], subset['total_users'], 
                          s=subset['total_users'] / 200,  # Mengatur ukuran bubble berdasarkan total_users
                          alpha=0.6)  # Transparansi
    # Menyimpan entri untuk legenda
    legend_handles.append(Line2D([0], [0], marker='o', color='w', 
                                   markerfacecolor=scatter.get_facecolor()[0], 
                                   markersize=10, label=season))

plt.title('Pengaruh Musim Terhadap Jumlah Total Pengguna Per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Total Pengguna')
plt.legend(handles=legend_handles, title="Musim")
plt.xticks(rotation=45)
plt.grid()

# Menampilkan koefisien korelasi di plot
plt.text(0.5, 0.5, f'Koefisien Korelasi: {correlation:.2f}', ha='center', va='center', transform=plt.gca().transAxes, fontsize=12, bbox=dict(facecolor='white', alpha=0.5))

plt.show()
print(f'Pengaruh Musim Terhadap Jumlah Total Pengguna Per Bulan: {correlation:.2f}')
```

```python
month_mapping = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}
season_month['month_num'] = season_month['month'].map(month_mapping)

# Visualisasi dengan bar plot
plt.figure(figsize=(12, 6))
bar_width = 0.15  # Lebar bar

for i, season in enumerate(season_month['season'].unique()):
    subset = season_month[season_month['season'] == season]
    plt.bar(subset['month_num'] + i * bar_width, 
            subset['total_users'], 
            width=bar_width, 
            alpha=0.7, 
            label=season)

plt.title('Pengaruh Musim Terhadap Jumlah Total Pengguna Per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Total Pengguna')
plt.xticks(list(month_mapping.values()), list(month_mapping.keys()))  # Menampilkan nama bulan
plt.legend(title="Musim")
plt.grid(axis='y')
plt.show()
```

### Pertanyaan 2: 
Apa hubungan antara jumlah pengguna kasual (casual) dan pengguna terdaftar (registered) di setiap bulan?

```python
monthly_users = df.groupby('month').agg({'casual': 'sum', 'registered': 'sum'}).reset_index()

# Mengonversi bulan menjadi angka untuk analisis lebih lanjut
month_mapping = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}
monthly_users['month_num'] = monthly_users['month'].map(month_mapping)

# Visualisasi hubungan antara pengguna kasual dan terdaftar
plt.figure(figsize=(10, 6))
sns.scatterplot(data=monthly_users, x='casual', y='registered', size='casual', sizes=(20, 200), alpha=0.5)

plt.title('Hubungan Antara Pengguna Kasual dan Terdaftar Per Bulan')
plt.xlabel('Jumlah Pengguna Kasual')
plt.ylabel('Jumlah Pengguna Terdaftar')
plt.grid()

# Menghitung dan menampilkan koefisien korelasi
correlation = monthly_users['casual'].corr(monthly_users['registered'])
plt.text(0.5, 0.5, f'Korelasi: {correlation:.2f}', fontsize=12, ha='center', transform=plt.gca().transAxes)

plt.show()

print(f'Korelasi antara jumlah pengguna kasual dan terdaftar: {correlation:.2f}')
```

```python
# Mengelompokkan pengguna kasual secara manual
def manual_group_casual(casual):
    if casual <= 500:
        return 'Very Low'
    elif 501 <= casual <= 1500:
        return 'Low'
    elif 1501 <= casual <= 3000:
        return 'Medium'
    else:
        return 'High'

# Mengelompokkan pengguna terdaftar secara manual
def manual_group_registered(registered):
    if registered <= 1000:
        return 'Very Low'
    elif 1001 <= registered <= 3000:
        return 'Low'
    elif 3001 <= registered <= 5000:
        return 'Medium'
    else:
        return 'High'

# Menerapkan pengelompokan
df['casual_group'] = df['casual'].apply(manual_group_casual)
df['registered_group'] = df['registered'].apply(manual_group_registered)

# Menghitung jumlah pengguna berdasarkan grup
casual_counts = df['casual_group'].value_counts()
registered_counts = df['registered_group'].value_counts()

# Membuat DataFrame untuk plotting
plot_data = pd.DataFrame({
    'Group': casual_counts.index,
    'Casual': casual_counts.values,
    'Registered': registered_counts.reindex(casual_counts.index, fill_value=0).values
})

# Membuat plot
plt.figure(figsize=(10, 6))

# Bar plot untuk pengguna kasual dan terdaftar
bar_width = 0.35
x = range(len(plot_data['Group']))

# Bar untuk pengguna kasual
plt.bar(x, plot_data['Casual'], width=bar_width, label='Pengguna Kasual', color='lightblue')

# Bar untuk pengguna terdaftar
plt.bar([p + bar_width for p in x], plot_data['Registered'], width=bar_width, label='Pengguna Terdaftar', color='lightgreen')

# Menambahkan label dan title
plt.title('Jumlah Pengguna Kasual dan Terdaftar Berdasarkan Pengelompokan Manual')
plt.xlabel('Grup Pengguna')
plt.ylabel('Jumlah Pengguna')
plt.xticks([p + bar_width / 2 for p in x], plot_data['Group'])
plt.legend()

# Menambahkan angka pada bar plot
for i in range(len(plot_data['Group'])):
    plt.text(i, plot_data['Casual'][i] + 20, plot_data['Casual'][i], ha='center', va='bottom')
    plt.text(i + bar_width, plot_data['Registered'][i] + 20, plot_data['Registered'][i], ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
